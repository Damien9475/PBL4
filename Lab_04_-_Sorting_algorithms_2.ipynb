## Exercice 4 : Sorting in linear time.
***

**Counting sort** is also called bucket sort or enumeration sort.

This sort is only applicable when the **list** to be sorted is **composed of integer values between two known bounds**. This constraint greatly reduces its scope of application.

<div style="border-color: #007bff; margin: 1.5625em auto; padding: 0 .6rem .8rem !important; overflow: hidden; page-break-inside: avoid;border-left: .2rem solid #007bff; border-radius: .1rem; box-shadow: 0 .2rem .5rem rgba(0,0,0,.05),0 0 .05rem rgba(0,0,0,.1); transition: color .25s,background-color .25s,border-color .25s;">

    <p style="padding: 0 1.4rem; margin-top: .4em; margin-bottom: 0; font-size: 1em;">
   
   However, when it is used, it is extremely efficient. Indeed, this efficiency is due to the fact that this sort is the only one that does not require any comparison between the different elements of the list to be sorted.
   
   </p>
</div>

The principle is based on constructing the histogram of the data, then scanning it in increasing order to reconstruct the sorted data.

Suppose in the following that we only have positive integer values. The principle of counting sort is as follows:
- Find the largest element $v$ in the set of elements.
- Construct an array of size $v+1$ with indices ranging from $0$ to $v$ and initialize the values of this array to 0.
- Traverse the set of values to be sorted, for each value $x$, increment the cell at index $x$.
- Traverse the array and construct the set of sorted elements by creating for each cell of the array as many values of the index as its content indicates.

<div style='color: forestgreen'>
    <h3>Example</h3>
</div>

If I have a list L of several grades ranging from 0 to 10 to be sorted in ascending order, counting sort will certainly be the best sort to use.

The principle is simple:
- _Step 1:_ Calculate the histogram
  - First, create a list `histogram` of 11 zeros that will be used for counting.
  - **Traverse** the list L to be sorted and take the first element i of our list L. Then add 1 to the index i of our `histogram` list used for **counting**.
  - Repeat this for all elements of the list.
- _Step 2:_ Deduce the sorted list from the histogram
  - Finally, **traverse all elements of the `histogram` list** to create our sorted list L_sorted.
  - Each time there is a non-zero value in the histogram list: add the value of the index of this histogram list to our L_sorted and add it as many times as the value of this cell.
